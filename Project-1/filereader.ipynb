```python
# Import libraries
from pathlib import Path
from datetime import datetime
import os
from collections import namedtuple

import pandas as pd
```

```python
# Create File namedtuple
File = namedtuple('File','name path size date_modified')
```

```python
# Create file list
files = []

# file path
p = Path('Z:/BU_DSC')

# Iterate through the path and find all .pdf files
# Save the list as a dataframe

for item in p.glob('**/*'):
	if item.suffix in (['.pdf']):
		name = item.name
		path = Path.resolve(item).parent
		size = item.stat().st_size
		modified = datetime.fromtimestamp(item.stat().st_mtime)

		files.append(File(name,path,size,modified))
```

```python
# Create dataframe from list
df = pd.DataFrame(files)
```
